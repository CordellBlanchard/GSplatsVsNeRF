## Convert Video into Images

```python
import cv2
import os

def video_to_images(video_path, output_folder, fps):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Get frames per second (fps) of input video
    input_fps = int(video_capture.get(cv2.CAP_PROP_FPS))

    # Calculate the frame interval to achieve the desired output fps
    frame_interval = round(input_fps / fps)

    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Read and save frames at the specified interval
    frame_count = 0
    success, frame = video_capture.read()
    while success:
        if frame_count % frame_interval == 0:
            output_file_path = os.path.join(output_folder, f"dim{frame_count // frame_interval:04d}.png")
            cv2.imwrite(output_file_path, frame)
        success, frame = video_capture.read()
        frame_count += 1

    # Release the video capture object
    video_capture.release()

if __name__ == "__main__":
    # Specify the input video file path
    input_video_path = "IMG_9451.MOV"

    # Specify the output folder for the image sequence
    output_folder = "/content/drive/MyDrive/Syn_nerf_DS/dimlight_bottle"

    # Specify the desired frames per second for the output image sequence
    desired_fps = 4

    # Convert video to image sequence
    video_to_images(input_video_path, output_folder, desired_fps)
```
